```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
P1=[0.1,0.6] 
P2=[0.15,0.71] 
P3=[0.08,0.9] 
P4=[0.16,0.85] 
P5=[0.2,0.3] 
P6=[0.25,0.5] 
P7=[0.24,0.1] 
P8=[0.3,0.2]

x=[P1,P2,P3,P4,P5,P6,P7,P8]
```

```python
x=np.array(x)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

```python
plt.scatter(P1[0],P1[1],c='red',label='P1')
plt.scatter(P2[0],P2[1],c='blue',label='P2')
plt.scatter(P3[0],P3[1],c='yellow',label='P3')
plt.scatter(P4[0],P4[1],c='green',label='P4')
plt.scatter(P5[0],P5[1],c='gold',label='P5')
plt.scatter(P6[0],P6[1],c='orange',label='P6')
plt.scatter(P7[0],P7[1],c='purple',label='P7')
plt.scatter(P8[0],P8[1],c='black',label='P8')
plt.title('Points in the dataset')
plt.legend()
plt.show()
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='cluster 1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='cluster 2')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='centroids')
plt.scatter(P6[0],P6[1],c='black',label='P6')
plt.title('Clustering')
plt.legend()
plt.show()
```
